```python
from ModelTester import *
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
signal = np.load('Data/Test_Data/Signal/101.npy')
annot_signal = signal = np.load('Data/Test_Data/Pred/101.npy')
annot_idx = ((wfdb.rdann('D:/ECG/mit/101', 'atr').sample / 360) * 250).astype(int)
```

```python
treshold_range = np.linspace(0, 0.5, 50)
true_pos, false_pos, false_neg, total_annot, total_target = score_data_treshold(signal,
                                                                               annot_signal,
                                                                               annot_idx,
                                                                               treshold_range, distance=75)
```

```python
n = total_target
plt.plot(treshold_range, np.asarray(true_pos)/n, label='True Positive rate')
plt.plot(treshold_range, np.asarray(false_pos)/n, label='False positive rate')
plt.plot(treshold_range, np.asarray(false_neg)/n, label='False negative rate')
plt.xlabel('Detection threshold')
plt.ylabel('Number of predictions/number of targets')
plt.legend()
plt.savefig('figs/Detection_threshold_curves.png')
#plt.show()
```
